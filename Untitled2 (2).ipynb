```python
import pandas as pd

# Load your file
df = pd.read_csv("PJME_hourly.csv")

# View data
print(df.head())
print(df.info())
```

```python
# Convert Datetime column to proper format
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Set Datetime as index
df.set_index('Datetime', inplace=True)

# Rename column for simplicity
df.rename(columns={'PJME_MW': 'Load'}, inplace=True)

# Check missing values
print(df.isnull().sum())

# Sort data (important for time series)
df = df.sort_index()

print(df.head())
```

```python
# Extract time-based features
df['hour'] = df.index.hour
df['day'] = df.index.day
df['month'] = df.index.month
df['dayofweek'] = df.index.dayofweek

print(df.head())
```

```python
# Lag feature (previous hour load)
df['lag_1'] = df['Load'].shift(1)

# Rolling average (last 24 hours)
df['rolling_mean_24'] = df['Load'].rolling(window=24).mean()

# Drop NaN values created by lag/rolling
df.dropna(inplace=True)

print(df.head())
```

```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestRegressor

# 🔷 STEP 1: Create features (IMPORTANT)
df['hour'] = df.index.hour
df['day'] = df.index.day
df['month'] = df.index.month
df['dayofweek'] = df.index.dayofweek

# 🔷 STEP 2: Lag + Rolling
df['lag_1'] = df['Load'].shift(1)
df['rolling_mean_24'] = df['Load'].rolling(window=24).mean()

# 🔷 STEP 3: Remove NaN rows (VERY IMPORTANT)
df = df.dropna()

# 🔷 STEP 4: Define features and target
features = ['hour', 'day', 'month', 'dayofweek', 'lag_1', 'rolling_mean_24']

X = df[features]
y = df['Load']

# 🔷 STEP 5: Time-based split
split = int(len(df) * 0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]

# 🔷 STEP 6: Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

print("Model trained successfully ✅")

# 🔷 STEP 7: Save threshold
threshold = df['Load'].quantile(0.9)
joblib.dump(threshold, "threshold.pkl")

# 🔷 STEP 8: Save model + features
joblib.dump(model, "energy_load_model.pkl")
joblib.dump(features, "features.pkl")
```

```python
# Make predictions
y_pred = model.predict(X_test)

print("Prediction completed ✅")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test.values[:200], label="Actual Load")
plt.plot(y_pred[:200], label="Predicted Load")

plt.legend()
plt.title("Actual vs Predicted Load")
plt.xlabel("Time")
plt.ylabel("Load (MW)")
plt.show()
```

```python
# Define threshold
threshold = df['Load'].quantile(0.9)

# Create peak labels
peak_actual = (y_test > threshold).astype(int)
peak_pred = (y_pred > threshold).astype(int)

print("Peak detection completed ✅")
```

```python
import joblib
import pandas as pd

# Load model + metadata
model = joblib.load("energy_load_model.pkl")
features = joblib.load("features.pkl")
threshold = joblib.load("threshold.pkl")

# Load dataset (IMPORTANT)
df = pd.read_csv("PJME_hourly.csv")
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)
df.rename(columns={'PJME_MW': 'Load'}, inplace=True)
df = df.sort_index()

print("⚡ Smart Energy Load Prediction CLI ⚡\n")

# User input (ONLY datetime)
date_input = input("Enter date (YYYY-MM-DD): ")
hour = int(input("Enter hour (0-23): "))

# Create full datetime
user_datetime = pd.to_datetime(f"{date_input} {hour}:00:00")

# 🔥 AUTO FEATURE ENGINEERING

# Lag_1 (previous hour load)
try:
    lag_1 = df.loc[user_datetime - pd.Timedelta(hours=1), 'Load']
except KeyError:
    print("❌ Previous hour data not available")
    exit()

# Rolling mean (last 24 hours)
try:
    rolling_window = df.loc[user_datetime - pd.Timedelta(hours=24): user_datetime - pd.Timedelta(hours=1)]
    rolling_mean_24 = rolling_window['Load'].mean()
except:
    print("❌ Not enough data for rolling mean")
    exit()

# Extract time features
day = user_datetime.day
month = user_datetime.month
dayofweek = user_datetime.dayofweek

# Create DataFrame (NO WARNING)
input_data = pd.DataFrame(
    [[hour, day, month, dayofweek, lag_1, rolling_mean_24]],
    columns=features
)

# Predict
predicted_load = model.predict(input_data)[0]

# Peak classification
is_peak = predicted_load > threshold

# Output
print("\n🔹 Predicted Load:", round(predicted_load, 2), "MW")

if is_peak:
    print("🔴 Peak Consumption Period")
else:
    print("🟢 Normal Consumption")
```
